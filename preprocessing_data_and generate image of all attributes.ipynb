```python
# pre-processing in text data
# step1 = remove punctuations
import string
string.punctuation
```

```python
def remove_punctuation(text):
    text = "".join([c for c in text if c not in string.punctuation])# remember this function.# .join() method takes an iterable of strings and combines them into a single string, using the specified separator between each element.
    return text
```

```python
data["text"] = data["text"].apply(lambda x: remove_punctuation(x)) # lambada function ( lambda arguments: expression )
data.head(1)
```

```python
import re # pattern matching and string manipulation
# removing unwanted characters : step2 (,,!)
data["text"] = data["text"].apply(lambda x: re.sub('[^A-Za-z0-9]+', " ", x)) #re.sub() function is used to perform regular expression-based substitution or replacement within a string
data["text"][0]     # [^..]-> negation , + -> If same condition bar bar ayega toh usko ek hi bar replace krenge. 
# In simpler terms, it removes any characters that are not letters (both uppercase and lowercase) or digits and replaces them with a space.
```

```python
# removing whitespaces --begining and trailing spaces : step3
data["text"] = data["text"].apply(lambda x: x.strip())  # strip function is a built-in string method that is used to remove specified characters (whitespace characters by default) from the beginning and end of a string.
data["text"][0]
```

```python
# replacing more than 1 spaces with single space : step4
data["text"] = data["text"].str.replace(r"\s\s+", " ") # the replace() method is a built-in string method that is used to create a new string by replacing all occurrences of a specified substring or character in a given string with another substring
data["text"][0]
```

```python
# tokenization: splitting up a larger body of text into smaller lines, words or even creating words for a non-English language.
def tokenize(text): # step 5
    tokens = re.split("\W+", text) # the split() method is a built-in string method used to split a string into a list of substrings based on a specified delimiter (separator).
    return tokens     # W+ ka matlab h jahan non-word aya wahan s split kar do.
   #It transforms raw text into a structured format that is suitable for analysis by ML models, enabling them to learn patterns and relationships within the language data.
```

```python
data["text"] = data["text"].apply(lambda x: tokenize(x.lower()))
data.head(1)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
wn = nltk.WordNetLemmatizer()
!unzip /usr/share/nltk_data/corpora/wordnet.zip -d /usr/share/nltk_data/corpora/
```

```python
# lemmatization : Lemmatization is technique used to reduce inflected words to their root word # step6 # better -> good is ex of leematization.
def lemmatization(text):
    text = [wn.lemmatize(word) for word in text]
    return " ".join(text)
```

```python
data["text"] = data["text"].apply(lambda x: lemmatization(x))
data.head(1)
```

```python
import os
from PIL import Image

# Assuming you have a DataFrame named 'data' with two columns 'Binary' and 'Name'
for index, row in new_data.iterrows():
    binary_string = row['Binary']
    image_name = row['image_id']  # Use the 'Name' column to name the image

    # Define the image width and height (adjust as needed)
    width = 28  # You can change this to your desired width
    height = 28

    # Ensure the binary string is padded with zeros if it's too short
    if len(binary_string) < width * height:
        binary_string = binary_string.ljust(width * height, '0')

    # Create an empty image with the specified width, height, and color mode
    mode = 'RGB'  # Change this if you have a different color mode
    image = Image.new(mode, (width, height))

    # Create an iterator over the binary string
    binary_iterator = iter(binary_string)

    # Iterate over the image pixels and set their values
    for y in range(height):
        for x in range(width):
            pixel_value = tuple(int(next(binary_iterator)) * 255 for _ in range(3))
            image.putpixel((x, y), pixel_value)

    # Define the folder where you want to save the images
    folder_path = 'images/'  # Change this to your desired folder path

    # Create the folder if it doesn't exist
    os.makedirs(folder_path, exist_ok=True)

    # Save each image with the 'image_name' as the filename
    image.save(os.path.join(folder_path, f'image_{image_name}.png'))

```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Set the path to your image folder
image_folder = '/Users/abhaydevraj/machine learning projects/blog attribute prediction/images'

# Set the path to your destination folders
training_folder = '/Users/abhaydevraj/machine learning projects/blog attribute prediction/training_folder'
validation_folder = '/Users/abhaydevraj/machine learning projects/blog attribute prediction/validation_folder'

# Create destination folders if they don't exist
os.makedirs(training_folder, exist_ok=True)
os.makedirs(validation_folder, exist_ok=True)

# Get the list of image files in the image folder
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Split the data into training and validation sets
train_files, val_files = train_test_split(image_files, test_size=0.2, random_state=42)

# Copy training set images
for file in train_files:
    source_path = os.path.join(image_folder, file)
    destination_path = os.path.join(training_folder, file)
    shutil.copy(source_path, destination_path)

# Copy validation set images
for file in val_files:
    source_path = os.path.join(image_folder, file)
    destination_path = os.path.join(validation_folder, file)
    shutil.copy(source_path, destination_path)

# Print the number of samples in the training and validation sets
print(f"Number of training samples: {len(train_files)}")
print(f"Number of validation samples: {len(val_files)}")
```
